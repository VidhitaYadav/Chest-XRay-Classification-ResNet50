```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input

BATCH_SIZE = 32
IMG_SIZE = (224, 224)

train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False,
    label_mode='int'
)

test_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False,
    label_mode='int'
)

# ResNet50 preprocessing (CRITICAL)
train_dataset = train_dataset.map(lambda x, y: (preprocess_input(x), y))
test_dataset  = test_dataset.map(lambda x, y: (preprocess_input(x), y))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test_true, y_test_pred)

plt.figure(figsize=(9,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

sensitivity = {}
specificity = {}

for i in range(len(class_names)): # Changed this line
    TP = cm[i, i]
    FN = np.sum(cm[i, :]) - TP
    FP = np.sum(cm[:, i]) - TP
    TN = np.sum(cm) - TP - FN - FP

    sensitivity[class_names[i]] = TP / (TP + FN) if (TP + FN) > 0 else 0
    specificity[class_names[i]] = TN / (TN + FP) if (TN + FP) > 0 else 0

print("\nSensitivity:")
for k, v in sensitivity.items():
    print(f"{k}: {v:.4f}")

print("\nSpecificity:")
for k, v in specificity.items():
    print(f"{k}: {v:.4f}")
```

```python
!pip install gradio matplotlib tensorflow --quiet

import gradio as gr
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt
from PIL import Image

# ---------------------------
# Load trained model
# ---------------------------
MODEL_PATH = "/content/drive/MyDrive/chest_xray_resnet50_best.keras"
model = tf.keras.models.load_model(MODEL_PATH)

# Enable dropout layers for Monte Carlo Dropout
for layer in model.layers:
    if "dropout" in layer.name.lower():
        layer.trainable = True

# Classes
class_names = [
    'Atelectasis', 'Cardiomegaly', 'Emphysema',
    'Lung Hernia', 'Normal', 'Pneumonia', 'Tuberculosis'
]

# Clinical notes
clinical_notes = {
    "Atelectasis": "Collapsed lung tissue",
    "Cardiomegaly": "Enlarged heart silhouette",
    "Emphysema": "Emphysema / lung damage",
    "Lung Hernia": "Lung herniation abnormality",
    "Normal": "No visible pathology",
    "Pneumonia": "Lung infection / consolidation",
    "Tuberculosis": "Bacterial lung infection"
}

# ---------------------------
# Grad-CAM Function
# ---------------------------
def get_gradcam(img_array, model, class_index, layer_name="conv5_block3_out"):
    grad_model = Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, class_index]
    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs), axis=-1)
    heatmap = np.maximum(heatmap, 0)
    heatmap /= (np.max(heatmap) + 1e-8)
    heatmap = np.uint8(255 * heatmap)
    heatmap = np.expand_dims(heatmap, axis=2)
    heatmap = tf.image.resize(heatmap, (224, 224)).numpy().astype(np.uint8)
    return heatmap

def overlay_gradcam(img, heatmap, alpha=0.4):
    heatmap = np.repeat(heatmap, 3, axis=2)
    heatmap_img = Image.fromarray(heatmap).convert("RGB")
    img = img.convert("RGB").resize((224,224))
    overlay = Image.blend(img, heatmap_img, alpha)
    return overlay

# ---------------------------
# Prediction Function
# ---------------------------
def predict_cxr(img: Image.Image):
    # Preprocess
    img_resized = img.resize((224,224))
    img_array = tf.keras.preprocessing.image.img_to_array(img_resized)
    img_array = preprocess_input(img_array)
    img_array_exp = np.expand_dims(img_array, axis=0)

    # Predict probabilities
    preds = model.predict(img_array_exp, verbose=0)[0]
    preds_percent = np.round(preds * 100, 2)

    # Monte Carlo Dropout
    mc_runs = 5
    mc_preds = []
    for _ in range(mc_runs):
        mc_preds.append(model.predict(img_array_exp, verbose=0)[0])
    mc_preds = np.array(mc_preds)
    uncertainty = float(np.std(mc_preds))

    # Prediction dictionary for Gradio
    label_scores = {class_names[i]: float(preds_percent[i]) for i in range(len(class_names))}

    # Top 3 predictions
    top3_idx = np.argsort(preds)[::-1][:3]
    top3_text = "\n".join([f"{class_names[i]}: {preds_percent[i]}%" for i in top3_idx])

    # Clinical notes
    notes = "\n".join([f"{cls}: {clinical_notes[cls]}" for cls in class_names])

    # Grad-CAM overlay for **top predicted class**
    heatmap = get_gradcam(img_array_exp, model, top3_idx[0])
    overlay_img = overlay_gradcam(img_resized, heatmap)

    return label_scores, f"Epistemic Uncertainty Score: {uncertainty:.6f}", notes, top3_text, overlay_img

# ---------------------------
# Gradio Interface
# ---------------------------
with gr.Blocks(title="Chest X-Ray Abnormality Prediction") as demo:
    gr.Markdown("<h1>Chest X-Ray Abnormality Prediction</h1>")
    gr.Markdown("Upload a CXR image to get predictions, uncertainty, top 3 diseases & Grad-CAM heatmap")

    with gr.Row():
        with gr.Column():
            img_input = gr.Image(type="pil", label="Upload Chest X-Ray")
        with gr.Column():
            label_output = gr.Label(num_top_classes=7, label="Prediction Results")
            uncertainty_output = gr.Textbox(label="Uncertainty Score")
            clinical_output = gr.Textbox(label="Clinical Notes")
            top3_output = gr.Textbox(label="Top 3 Predicted Diseases")
            gradcam_output = gr.Image(label="Grad-CAM Overlay")

    btn = gr.Button("Predict")
    btn.click(
        predict_cxr,
        inputs=img_input,
        outputs=[label_output, uncertainty_output, clinical_output, top3_output, gradcam_output]
    )

# Launch app
demo.launch(debug=True)
```
